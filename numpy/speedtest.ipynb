```python
#using python lists

import numpy as np
import time

size = 1_000_000 #for better readability
l3 = list(range(size))
l4 = list(range(size))
start = time.time()
add = [i + j for i, j in zip(l3, l4)]
print(add)
end = time.time()
print("List took: ", end - start)
```

```python
#using numpy arrays

import numpy as np
import time

size = 1_000_000 #for better readability
l3 = np.array(list(range(size)))
l4 = np.array(list(range(size)))
start = time.time()
add = l3 + l4
print(add)
end = time.time()
print("List took: ", end - start)
```

```python
# Python list (loop-based)
start = time.time()
list_squares = [x ** 2 for x in l3]
end = time.time()
print("List comprehension took:", end - start)

# NumPy (vectorized)
start = time.time()
numpy_squares = l3 ** 2
end = time.time()
print("NumPy vectorization took:", end - start)
```

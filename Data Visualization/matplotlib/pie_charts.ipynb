Pie Charts in Matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]

plt.title('Programming language usage')
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True, 'linestyle': 'dashed'})
plt.show()
```

```python
explode = (0, 0.4, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=100, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True, 'linestyle': 'dashed'},
        shadow=True)
plt.show()
```

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```

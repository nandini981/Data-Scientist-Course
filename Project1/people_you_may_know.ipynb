```python
import json

def load_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data


def find_people_you_may_know(user_id, data):
    user_friends = {}

    # Build user -> friends mapping
    for user in data['users']:
        user_friends[user["id"]] = set(user['friends'])

    # If user does not exist
    if user_id not in user_friends:
        return []

    direct_friends = user_friends[user_id]
    potential_friends = {}

    # Look at friends of friends
    for friend_id in direct_friends:
        for mutual in user_friends.get(friend_id, []):
            if mutual != user_id and mutual not in direct_friends:
                # âœ… COUNT ONLY ONCE
                potential_friends[mutual] = potential_friends.get(mutual, 0) + 1

    # Sort by number of mutual friends (descending)
    suggested_friends = sorted(
        potential_friends.items(),
        key=lambda x: x[1],
        reverse=True
    )

    return [friend_id for friend_id, _ in suggested_friends]


# Load the data
data = load_json('massive_data.json')
user_id = 10
suggested_friends = find_people_you_may_know(user_id, data)

print(f"People you may know for user {user_id}: {suggested_friends}")
```

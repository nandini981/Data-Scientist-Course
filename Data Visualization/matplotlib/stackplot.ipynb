Stack Plots in Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
activities = ['eat', 'sleep', 'work', 'play']
time_spent = [2, 7, 8, 7]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.figure(figsize=(5, 5))
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True, 'linestyle': 'dashed'})

plt.title('Time spent on activities')
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt
 
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors=colors)
plt.title('Daily Activities')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
